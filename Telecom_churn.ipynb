**Graph between senior citizen and non senior citizen (in barchart form)**

This code calculates the percentage distribution of churn for senior and non-senior customers and visualizes it using a stacked bar chart for easy comparison.

```python
counts = df.groupby(['SeniorCitizen', 'Churn']).size().unstack()

# Convert to percentage
percent = counts.div(counts.sum(axis=1), axis=0) * 100

# Define colors using seaborn's palette
colors = sns.color_palette("mako", n_colors=2)

# Plot
ax = percent.plot(
    kind='bar',
    stacked=True,
    figsize=(7,5),
    color=colors # Use the 'color' argument with the defined colors
)

# Add labels
for c in ax.containers:
    ax.bar_label(c, fmt='%.1f%%', label_type='center', fontsize=8)

ax.set_title('Senior Citizen vs Churn (%)')
ax.set_xlabel('Senior Citizen')
ax.set_ylabel('Percentage')
ax.set_xticklabels(['Not Senior', 'Senior'], rotation=0)
ax.legend(title='Churn')

plt.tight_layout()

plt.show()
```

**Graph between senior citizen and non senior citizens(in pie chart form)**

This code creates side-by-side pie charts to compare churn distribution between senior and non-senior customers.

```python
counts = df.groupby(['SeniorCitizen', 'Churn']).size().unstack()

colors = sns.color_palette("mako", n_colors=2)

fig, axes = plt.subplots(1, 2, figsize=(7,3))

axes[0].pie(
    counts.loc["No"],
    labels=counts.columns,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors
)
axes[0].set_title('Not Senior Citizens')

axes[1].pie(
    counts.loc["Yes"],
    labels=counts.columns,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors
)
axes[1].set_title('Senior Citizens')

plt.tight_layout()
plt.show()
```

**Creating subplots on the basis of columns**

```python
df.columns.values
```

This code generates multiple countplots for all categorical features, showing their distribution split by churn using a consistent grid layout for easy comparison.

```python
sns.set_style("whitegrid")
# List of categorical columns
cat_cols = [
    'gender', 'SeniorCitizen', 'Partner', 'Dependents',
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies',
    'Contract', 'PaperlessBilling', 'PaymentMethod', 'Churn'
]

# Subplot configuration
n_cols = 3
n_rows = (len(cat_cols) + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, 5 * n_rows))
axes = axes.flatten()

# Create countplots
for i, col in enumerate(cat_cols):
    sns.countplot(
        data=df,
        x=col,
        hue='Churn',
        palette='mako',
        ax=axes[i],
        edgecolor='black' # Added a comma here
    )
    axes[i].bar_label(axes[i].containers[0])
    axes[i].set_title(col)
    axes[i].tick_params(axis='x', rotation=45)

# Remove empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

Clear differences are visible across variables such as Contract type, InternetService, and PaymentMethod, indicating their strong association with churn.
Customers on month-to-month contracts and those using electronic check payments exhibit higher churn rates.
